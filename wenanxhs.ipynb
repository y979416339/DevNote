```
# 乱序，划分训练集和测试集
import random

nplst=[]
with open("/content/drive/MyDrive/shuyusentence.txt", 'rt') as handle:
    for data in handle.readlines():
        data = data.strip('\n')
        if data != "":
#         print(data)
            nplst.append(data)

# 数据集一共3904个
random.shuffle(nplst)
# [训练集, 测试集]
pos = int(len(nplst) *.7)
# print(pos)
parts = nplst[:pos], nplst[pos:]
train_sy=parts[0] #训练集共2732个
test_sy=parts[1] #测试集共1172个
# print(train_sy)
# print(len(train_sy))
# print(test_sy)
# print(len(test_sy))
# print(len(nplst))
# nplst

#将训练集和测试集的数据写入文件

f1=open("/content/drive/MyDrive/train_sy.txt","w")
for line in train_sy:
    f1.write(line+'\n')
f1.close()

f2=open("/content/drive/MyDrive/test_sy.txt","w")
for line in test_sy:
    f2.write(line+'\n')
f2.close()
```
